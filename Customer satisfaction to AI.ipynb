# Categorical variable analysis help to understand the distribution, relation between another variables and overall significance

# 1. Frequency Distribution

```python
for x in df.columns:
    print(f"value counts for {x}:")
    print(df[x].value_counts())
    print(f"Normalize {x}:")
    print(df[x].value_counts(normalize=True))
    print()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y=df['Country'],order=df['Country'].value_counts().index,palette="viridis")
```

```python
num_cols = len(df.columns)
fig, axes = plt.subplots(nrows=1, ncols=num_cols, figsize=(5 * num_cols, 5), constrained_layout=True)
if num_cols == 1:
    axes = [axes]


for ax, col in zip(axes, df.columns):
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, ax=ax, palette="viridis")
    ax.set_title(f"Value Counts for {col}")
    ax.set_xlabel("Count")
    ax.set_ylabel("Categories")

plt.show()
```

```python
for x in df.columns:
    df[x].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title('Category Proportions')
    plt.show()
```
